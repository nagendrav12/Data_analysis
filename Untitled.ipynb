# Visualization

```python
# Define custom color palette
sns.set_palette(["green", "blue", "pink", "yellow"])
sns.set_style("whitegrid")
# Create bubble chart
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='age', y='charges', size='bmi', sizes=(20, 200), alpha=0.7, hue=df['region'].head(200))
plt.title('Age vs. Charges (Bubble Chart)', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Charges', fontsize=14)
plt.legend(title='Region')
plt.show()
```

# Certainly! Let's break down the code:

1. **Importing Libraries**: 
   - We import the necessary libraries: `seaborn` and `matplotlib.pyplot`. Seaborn is a statistical data visualization library built on top of Matplotlib.

2. **Customizing Visualization**:
   - `sns.set_palette(["green", "blue", "pink", "yellow"])`: This sets a custom color palette for the plot. The colors defined in the list will be used for different categories in the plot.
   - `sns.set_style("whitegrid")`: This sets the style of the plot. Here, we're using the "whitegrid" style, which shows gridlines on a white background.

3. **Creating the Bubble Chart**:
   - `plt.figure(figsize=(10, 8))`: This sets the figure size for the plot.
   - `sns.scatterplot(...)`: This creates the scatter plot. We're using `sns.scatterplot` from Seaborn, which allows us to create scatter plots with additional features like size and color encoding. 
     - `data=df.head(200)`: This specifies the DataFrame to use for plotting. We're using only the first 200 rows of the DataFrame `df`.
     - `x='age', y='charges'`: This sets the x and y variables for the plot.
     - `size='bmi'`: This encodes the size of the markers based on the 'bmi' column.
     - `sizes=(20, 200)`: This sets the range of sizes for the markers.
     - `alpha=0.7`: This controls the transparency of the markers.
     - `hue='region'`: This encodes the color of the markers based on the 'region' column.
   - `plt.title(...)`, `plt.xlabel(...)`, `plt.ylabel(...)`: These functions set the title, x-axis label, and y-axis label for the plot, respectively.
   - `plt.legend(title='Region')`: This adds a legend to the plot with the title 'Region', which indicates the meaning of the colors used in the plot.
   - `plt.show()`: This displays the plot.

Overall, this code creates a bubble chart visualizing the relationship between age and charges, with markers sized based on BMI and colored based on the region. The plot is customized with a specific color palette and style.

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Certainly! Here's a condensed explanation of the code:

1. **Importing Libraries**: We import Seaborn and Matplotlib for visualization.

2. **Creating the Heatmap**: 
   - We use Seaborn's `heatmap` function to create a heatmap of the correlation matrix of the dataset.
   - `annot=True` adds correlation values to the heatmap cells.
   - `cmap='coolwarm'` sets the color palette to 'coolwarm'.
   - `fmt=".2f"` formats the annotation values with two decimal places.

3. **Interpreting the Heatmap**: 
   - Each cell represents the correlation coefficient between two variables.
   - Darker colors (closer to red) indicate positive correlation, lighter colors (closer to blue) indicate negative correlation.
   - Values close to zero indicate little to no correlation.

Overall, the heatmap visually represents the correlations between variables in the dataset.
